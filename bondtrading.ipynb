```python
import pandas as pd
import numpy as np
import string
import random
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (15,3)

def isin(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
def vol():
    return random.randint(200000,10000000)
def month():
    return random.randrange(1,12)
isin()
vol()
month()
data = []
for i in range(1,100):
    data.append({"isin":isin(size=3,chars='ABC'),"vol":vol(),"month":month()})
bdf = pd.DataFrame(data)
bgp1 = bdf[(bdf.month == 1)].groupby(['month','isin']).agg(['sum'])
bgp2 = bdf[(bdf.month == 1)].groupby(['month','isin']).agg(['sum'])
bgp3 = bdf[(bdf.month == 1)].groupby(['month','isin']).agg(['sum'])

bgp1.plot(kind='bar')
bgp2.plot(kind='bar')
bgp3.plot(kind='bar')
```
